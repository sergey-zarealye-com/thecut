###Наивная реализация LHS 

```python
class LhsSimple:
    
    def __init__(self, R=64, D=1024):
        self.R = R #размерность хэша
        self.D = D #размерность вектора данных
        np.random.seed(seed=1968)
        self.rnd_vectors = np.random.randint(2, size=(R, D))*2 - 1
        
    def binstr_hash(self, x):
        dot_ = np.dot(self.rnd_vectors, x)
        bin_ = (np.sign(dot_) + 1) / 2
        return ''.join(str(int(e)) for e in bin_)
    
    def int_hash(self, x):
        return int(self.binstr_hash(x), 2)
```

### Итератор по фильму позволяет получать кадры, упакованные в мини-батчи.
Входные данные:
-    fname - путь к файлу видео.
-    scene_cuts (опционально) - список границ сцен в миллисекундах. Если он задан, мини-батчи формируются с учетом этого, т.е. в мини-батч попадают только кадры из одной сцены, не более CUT_MAX_FRAMES.
-    CUT_MAX_FRAMES не должен быть больше 10 (ограничение реализации GoogLeNet) игнорируется если scene_cuts не задан!
-    STEP_MSEC - интервал выбороки кадров из фильма в миллисекундах
-    MAXTIME - предел длительности фильма в миллисекундах; 0 - нет предела (весь фильм)
-    BUFFSIZE - максимальный размер минибатча

```python
def frame_iterator(fname, scene_cuts=None, CUT_MAX_FRAMES=10, STEP_MSEC = 200, 
                   MAXTIME = 120000, BUFFSIZE = 10):
    buff_cnt = 0
    buff = []
    pos_list = []
    cap = cv2.VideoCapture(fname)
    current_cut = 0
    if cap.isOpened():
        frame_msec = 0.
        
        try:
            while True:
                ret = cap.grab()
                if ret > 0:
                    msec = cap.get(cv2.CAP_PROP_POS_MSEC) 
                    pos_frames = cap.get(cv2.CAP_PROP_POS_FRAMES)
                    pos_avi_ratio = cap.get(cv2.CAP_PROP_POS_AVI_RATIO)
                    if MAXTIME > 0 and msec > MAXTIME:
                        print "MAXTIME reached", msec                        
                        break
                    if msec - frame_msec >= STEP_MSEC:
                        if scene_cuts is None:
                            ret, frame = cap.retrieve()
                            frame_msec = msec   
                            buff_cnt += 1
                            buff.append(frame / 255.)
                            pos_list.append({'pos_sec': msec / 1000.,
                                            'pos_frames': pos_frames,
                                            'pos_avi_ratio': pos_avi_ratio})
                            if buff_cnt >=BUFFSIZE :    
                                yield pos_list, buff
                                buff_cnt = 0
                                buff = []
                                pos_list = []
                        else:
                            if buff_cnt < CUT_MAX_FRAMES and (
                                    current_cut >= len(scene_cuts) or                                    
                                    msec < scene_cuts[current_cut] ):
                                ret, frame = cap.retrieve()
                                frame_msec = msec   
                                buff_cnt += 1
                                buff.append(frame / 255.)
                                pos_list.append({'pos_sec': msec / 1000.,
                                                'pos_frames': pos_frames,
                                                'pos_avi_ratio': pos_avi_ratio})
                                
                            if current_cut < len(scene_cuts) and msec >= scene_cuts[current_cut]:
                                current_cut += 1
                                if len(pos_list):
                                    yield pos_list, buff
                                    buff_cnt = 0
                                    buff = []
                                    pos_list = []
                else:
                    break
        except GeneratorExit:
            pass
        except Exception:
            print traceback.format_exc()
        cap.release()
        cv2.destroyAllWindows()
        if len(pos_list) and (scene_cuts is None):
            yield pos_list, buff
        elif len(pos_list) and (scene_cuts is not None):
            yield pos_list, buff
```

###Получаем кадры из фильма, рассчитываем вектор признаков и по расстоянию между векторами определяем разрывы сцен

```python
def segment_film(fname, 
                   CUT_MAX_FRAMES=10, 
                   STEP_MSEC = 320, 
                   MAXTIME = 60000,
                   BUFFSIZE = 10,
                   DIST_THRE = 0.01,
                   SCENE_THR = 0.15,
                   NRELEASE = 2):
    RELEASE = STEP_MSEC * NRELEASE / 1000.
    #FILTR = np.array([-0.04340667,-0.46344629,0.,0.46344629,0.04340667])             #sigma 0.7
    #FILTR = np.array([-0.12689981,-0.21918588,0.,0.21918588,0.12689981])             #sigma 1.1
    FILTR = np.array([-0.01428036, -0.46054231, 0., 0.46054231, 0.01428036])          #sigma 0.6.1
    NFILTR = len(FILTR)
    
    prev_feature_vect = None
    scene_distance = 0
    scene_distances = [0.]
    time_keys = []
    
    o_scene_distances = []
    o_time_keys = []
    
    last_detected = 0
    lastsofar = 0
    online_scene_dists = [0.]
    online_scene_scene_keys = []
    scene_fvectors = []

    for sec_list, inputs in frame_iterator(fname, CUT_MAX_FRAMES=10, 
                                           STEP_MSEC = STEP_MSEC, 
                                           MAXTIME = MAXTIME,
                                           BUFFSIZE = BUFFSIZE):
        time_keys.extend(sec_list)

        # Classify.
        start = time.time()
        #predictions, features = classifier.predict(inputs, False, blob = 'pool5/7x7_s1')
        predictions = classifier.predict(inputs, False)
        features = classifier.blobs['pool5/7x7_s1'].data
        for i in range(len(predictions)):

            feature_vect = features[i, :, :, :].flatten()
            if prev_feature_vect is not None:
                scene_distance = np.linalg.norm(
                    prev_feature_vect/np.linalg.norm(prev_feature_vect) - 
                    feature_vect/np.linalg.norm(feature_vect)
                )
                if scene_distance < DIST_THRE:
                    scene_distance = 0.
                scene_distances.append(scene_distance)

            prev_feature_vect = np.array(feature_vect)

            online_scene_dists.append(scene_distance)    
            
            if len(online_scene_dists) > NFILTR:
                online_scene_dists.pop(0)
                buff = np.array(online_scene_dists)
                resp = np.dot(buff, FILTR)
                sec_list[i]['pos_sec'] -= (NFILTR/2-1)*STEP_MSEC/1000.
                sec_list[i]['pos_frames'] -= NFILTR/2-1
                
                o_time_keys.append(sec_list[i])
                o_scene_distances.append(resp)
                if resp >= SCENE_THR and sec_list[i]['pos_sec'] > last_detected + RELEASE:
                    online_scene_scene_keys.append(sec_list[i])
                    last_detected = sec_list[i]['pos_sec']
                    scene_fvectors = []

    #return online_scene_scene_keys, (scene_distances, time_keys)  
    return online_scene_scene_keys, (o_scene_distances, o_time_keys)   
```

###Оценка точности сегментации

- gt_frames образцовые номера кадров (начало сегментов) 
- our_frames детектированные номера кадров (начало сегментов) 
- TOLERANCE допустимая погрешность в определении номера ключевого кадра

Выходные значения:
    
- TP true positives число совпадений найденных номеров с образцовыми +- TOLERANCE
- FP false positives число кадров в our_frames, не совпавших с образцовыми
- FN false negatives число кадров в gt_frames, не совпавших с найденными.

```python
def evaluate_cuts(gt_frames, our_frames, TOLERANCE=5):
    TP = 0
    total_err = 0.0
    start_idx = 0
    for gtframe in gt_frames:
        for our_idx in xrange(start_idx, len(our_frames)):
            err = abs(gtframe - our_frames[our_idx])
            if err <= TOLERANCE:
                TP += 1
                start_idx = our_idx + 1
                total_err += err
                break
    FP = len(our_frames) - TP
    FN = len(gt_frames) - TP
    return TP, FP, FN, total_err / TP
```

```python
totals = {}
for scene_thr in (80., 85., 90., 95., 100.):
    TP = 0
    FP = 0
    FN = 0
    for res in results1:
        if res['SCENE_THR'] == scene_thr:
            TP += res['TP']
            FP += res['FP']
            FN += res['FN']
    P = 1.0 * TP / (TP + FP)
    R = 1.0 * TP / (TP + FN)
    ACC = 1.0 * TP / (TP + FP + FN)
    F1 = 2.0 * R * R / (P + R)
    totals[scene_thr] = {'TP':TP, 'FP':FP, 'FN':FN, 'P':P, 'R':R, 'ACC':ACC, 'F1': F1}

precision = [(k, totals[k]['P']) for k in totals]
precision.sort(key=lambda o: o[0])

recall = [(k, totals[k]['R']) for k in totals]
recall.sort(key=lambda o: o[0])

accuracy = [(k, totals[k]['ACC']) for k in totals]
accuracy.sort(key=lambda o: o[0])

f1 = [(k, totals[k]['F1']) for k in totals]
f1.sort(key=lambda o: o[0])
```
